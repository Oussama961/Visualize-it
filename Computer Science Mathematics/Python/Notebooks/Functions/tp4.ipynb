### Exercice 1:

Ecrire un programme pour gérer les informations personelles d'un ensemble de contactes. chaques contact est représenté par les informations suivants:
* Contact:
    - `nom`       (`list`/Chaine de caracteres)
    - `prenom`    (`list`/Chaine de caracteres)
    - `email`     (`list`/Chaine de caracteres)
    - `telephone` (`list`/Chaine de caracteres)

* 1/ Un enregistrement (ou structure) pour représenter un contact.
* 2/ Une foncion `ajouter_contacte()` pour creer un contact à partir des données saisir par l'utilisateur.
* 3/ Le programme principale qui saisit N contacts dans un tableau, et envoque
* 4/ Programme principale qui saisit N contacts dans un tableau, et invoque les différents fonctions.

```python
from dataclasses import dataclass

@dataclass
class Contact:
    nom: str
    prenom: str
    email: str
    telephone: str

def contact() -> Contact:
    nom = input("Nom: ")
    prenom = input("Prenom: ")
    email = input("email: ")
    telephone = input("Telephone: ")
    c = Contact(nom, prenom, email, telephone)
    return c

def affichage_contact(c: Contact):
    print(f"Le nom: {c.nom}")
    print(f"Le nom: {c.prenom}")
    print(f"Le nom: {c.email}")
    print(f"Le nom: {c.telephone}")


def process():
    Dis = str(input("Ajouter un contact (Click 1) / Affichage d'un contact (2).\nProcessing... :"))
    if Dis == 1:
        data = contact()
        T.append(data)
        dis = int(input("Pour ajouter un autre contact, click 'Y/y': "))
        if dis == ('Y' or 'y'):
            process()
        else:
            return T
    elif Dis == 2:
        affichage_contact(T)
        if T == []:
            print("La list est vide, ajouter un contact pour afficher")
            return process()
    else: return

    
if __name__ == "__main__":
    T = []
    process()
```

```python
from dataclasses import dataclass

@dataclass
class Voiture:
    immatriculation: str
    modele: str
    annee: int
    disponible: bool = True

@dataclass
class Client:
    nom: str
    prenom: str
    carte_nationale: str
    photo: str
    email: str
    telephone: str

@dataclass
class Location:
    immatriculation: str
    clien_id: str
    date_debut: str
    date_fin: str


def saisie_voiture() -> Voiture:
    imm = input("Immatriculation : ")
    mod = input("Modele : ")
    an  = int(input("Annee : "))
    return Voiture(imm, mod, an)

def affichage_voiture(v: Voiture):
    if v.disponible:
      status = "disponible"
    else:
      status = "pas disponible"
    print(f"{v.immatriculation} , {v.modele} ({v.annee}) : {status}")


def saisie_client() -> Client:
    nom  = input("Nom : ")
    pre  = input("Prénom : ")
    mail = input("Email : ")
    tel  = input("Telephone : ")
    return Client(nom, pre, mail, tel)

def affichage_client(c: Client):
    print(f"{c.carte_id} , {c.nom} {c.prenom} - {c.telephone} [{c.photo}]")

def affichage_location(loc: Location):
    print(f"Voiture {loc.immatriculation} louée par {loc.client_email} du {loc.date_debut} au {loc.date_fin}")


def charger(filename: str):
  try:
    with open(filename, "r") as f:
      f.read()
    return f
  except FileNotFoundError:
    return "[]"

def sauvegarder(objets: List, filename: str):
    with open(filename, "w") as f:
        for i in range(len(objets)):
            ch = f"{objets[i].nom}; {objets[i].prenom}; {objets[i].email}; {objets[i].telephone}; "
            f.write(ch)

        

def load_dataclasses(filename: str, cls):
    data = charger(filename)
    c_class = []
    for d in data:
      c_class.append(d)
    return c_class


def louer_voiture(voitures: List[Voiture], clients: List[Client], locations: List[Location]):
    imm = input("Immatriculation de la voiture à louer : ")
    for v in voitures:
      if (v.immatriculation == imm and v.disponible):
         print("Voiture disponible")
         email = input("Email du client : ")
         break
      else:
        print("Voiture non disponible ou inexistante.")
        return
      
    carte_id = input("ID du client : ")
    for c in clients:
      if c.carte_id == carte_id :
        print(f"Client {c.nom}, {c.prenom}")
      else:
        print("Client pas trouve.")
        return
    date_deb = input("Entrer les dates: \n debut: ")
    date_fin = input("\n fin : ")
    loc = Location(imm, carte_id, date_deb, date_fin)
    locations.append(loc)
    v.disponible = False
    print(f"Voiture {imm} louee a {email} depuis {date_deb}.")

def retourner_voiture(voitures: Voiture, locations: Location):
    imm = input("Immatriculation de la voiture à retourner : ")
    for l in location:
      if l.immatriculation == imm:
        print(f"Immatricule est trouvee")
      else:
        print("Aucune location en cours pour cette voiture.")
        return

    for v in voitures:
      if v.immatriculation == imm:
        v.disponible = True
      else:
        print("Aucune location en cours pour cette voiture.")
        return
    print(f"Voiture {imm} retournée.")


def main():
    voitures: Voiture   = load_dataclasses("voitures.txt", Voiture)
    clients: Client     = load_dataclasses("clients.txt", Client)
    locations: Location = load_dataclasses("locations.txt", Location)

    menu = """
1. Ajouter une voiture: 
2. Lister les voitures: 
3. Ajouter un client: 
4. Lister les clients: 
5. Louer une voiture: 
6. Retourner une voiture: 
7. Lister les locations: 
8. Sauvegarder et quitter: 
"""
    while True:
        print(menu)
        choix = input("Choix :")
        if choix == "1":
            v = saisie_voiture()
            voitures.append(v)
        elif choix == "2":
            for v in voitures:
                affichage_voiture(v)
        elif choix == "3":
            c = saisie_client()
            clients.append(c)
        elif choix == "4":
            for c in clients:
                affichage_client(c)
        elif choix == "5":
            louer_voiture(voitures, clients, locations)
        elif choix == "6":
            retourner_voiture(voitures, locations)
        elif choix == "7":
            for l in locations:
                affichage_location(l)
        elif choix == "8":
            sauvegarder(voitures,   "voitures.txt")
            sauvegarder(clients,    "clients.txt")
            sauvegarder(locations,  "locations.txt")
            print("Données sauvegardees. Au revoir!")
            break
        else:
            print("Choix invalide, reessayez.")

if __name__ == "__main__":
    main()
```
